Pandas and Numpy


```python
import pandas as pd

# create a dictionary
data= {'Name':['Aayusha','Aaradhya','Priyansh','Bhuvika','Amishi'],
       'Age':[25,30,35,15,18],
       'City':['New York','London','Paris','Delhi','Seoul']}

# create a dataframe from dictionary
df= pd.DataFrame(data)

print(df)
```

```python
import pandas as pd
import numpy as np

# creating empty series
ser= pd.Series()
print("Pandas Series:", ser)

# simple array
data= np.array(['a','a','y','u','s','h','a'])

ser= pd.Series(data)
print("Pandas Series:\n", ser)
```

```python
import pandas as pd

# create a 2D list
data= [['Aayusha', 25, 'New York','Kia','Iphone'],
       ['Aaradhya',30,'London','Tesla','Iphone'],
       ['Priyansh',35,'Paris','Mahindra','One Plus'],
        ['Bhuvika',15,'Delhi','Hyundai','Samsung']]

# create a dataframe from list
df= pd.DataFrame(data, columns=['Name','Age','City','Car','Phone'])

print(df)
```
