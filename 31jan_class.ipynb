```python
# find mean, median, mode and standard deviation of dataset given by using for loop without using in built function
import pandas as pd
from google.colab import files
# Upload the file
uploaded = files.upload()
# Load data from the uploaded Excel file
df1 = pd.read_csv("heart1.csv")

# Select a specific column
column_name = 'trestbps'
data = list(df1[column_name].dropna())  # Convert column to list and remove NaN values

import math

# 1. Calculate Mean
sum_values = 0
count = 0
for num in data:
    sum_values += num
    count += 1

mean_value = sum_values / count
print(f"Mean: {mean_value}")

# 2. Calculate Median
sorted_data = sorted(data)
n = len(sorted_data)

if n % 2 == 1:
    median_value = sorted_data[n // 2]  # Odd length, middle element
else:
    median_value = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2  # Even length, average of middle two

print(f"Median: {median_value}")

# 3. Calculate Mode
frequency = {}  # Dictionary to store frequency of elements
for num in data:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

max_freq = 0
modes = []

for key, value in frequency.items():
    if value > max_freq:
        max_freq = value
        modes = [key]  # Reset mode list with new highest frequency element
    elif value == max_freq:
        modes.append(key)  # Add to mode list if frequency matches max_freq

print(f"Mode: {modes}")

# 4. Calculate Variance
sum_squared_diff = 0

for num in data:
    sum_squared_diff += (num - mean_value) ** 2  # Squared difference from mean

variance = sum_squared_diff / count  # Population variance

# 5. Calculate Standard Deviation (Square Root of Variance)
std_deviation = math.sqrt(variance)

print(f"Standard Deviation: {std_deviation}")
```

```python
# data normalisation without using in buit functions

import pandas as pd
from google.colab import files
uploaded = files.upload()
df1 = pd.read_csv("heart1.csv")

# Select a specific column
column_name = 'oldpeak'
data = list(df1[column_name].dropna())  # Convert column to list and remove NaN values

# 1. Find Min and Max
min_value = data[0]
max_value = data[0]

for num in data:
    if num < min_value:
        min_value = num
    if num > max_value:
        max_value = num

# 2. Apply Min-Max Normalization
normalized_data = []

for num in data:
    norm_value = (num - min_value) / (max_value - min_value)  # Min-Max formula
    normalized_data.append(norm_value)

# Print Results
print(f"Original Data: {data}")
print(f"Normalized Data: {normalized_data}")
```
